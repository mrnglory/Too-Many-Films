```python
from bs4 import BeautifulSoup
import urllib.request
import pandas as pd
```

```python
df = pd.DataFrame(columns=["title", "genre", "year", "synopsis", "link", "img"])
count = 0
```

```python
for num in range(1, 7167):
    url = "https://www.4flix.co.kr/board/netflix/" + str(num)
    with urllib.request.urlopen(url) as url:
        try:
            doc = url.read()
            soup = BeautifulSoup(doc, "html.parser")
            
            title_year = soup.find_all("h1")[2].text.strip()
            
            title = title_year[:-6]
            year = title_year[-5:-1]
            genre = soup.find_all("h3")[1].text.strip()
            synopsis = soup.find_all("p")[0].text.strip()
            link = soup.find_all("a")[41]["title"]
            
            style_tag = list(soup.find_all("style")[3])
            left_index = 316
            right_index = style_tag[0].index(")")
            
            if style_tag[0][left_index:right_index] is not None:
                img = style_tag[0][left_index:right_index]
            else:
                img = ""
            
            if synopsis[0:2] == "수상":
                synopsis = soup.find_all("p")[1].text.strip()
            else:   
                synopsis = soup.find_all("p")[0].text.strip()
            
            df.loc[count] = [title, genre, year, synopsis, link, img]
            count += 1
        except:
            pass
```
